Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

Load Dataset

```python
data = pd.read_csv(r"C:\Users\HP\Downloads\Cas.csv\Cas.csv")

print(data.head())
```

 Shape and Columns

```python
print("Dataset Shape:", data.shape)

print("\nColumns:")
print(data.columns)
```

Missing Values


```python
print(data.isnull().sum())
```

Summary Statistics

```python
print(data.describe())
```

Casualty Severity Distribution

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Casualty_Severity',
    data=data,
    hue='Casualty_Severity',
    palette='viridis',
    legend=False
)

plt.title("Casualty Severity Distribution")

plt.xlabel("Casualty Severity")

plt.ylabel("Count")

plt.show()
```

 Age Distribution


```python

plt.figure(figsize=(10,5))

sns.histplot(
    data['Age_of_Casualty'],
    bins=20,
    color='purple'
)

plt.title("Age Distribution of Casualties")

plt.xlabel("Age")

plt.ylabel("Frequency")

plt.show()
```

Interpretation:

Most casualties belong to young and middle-aged groups.

 Gender Distribution


```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Sex_of_Casualty',
    data=data,
    hue='Sex_of_Casualty',
    palette='magma',
    legend=False
)

plt.title("Gender Distribution of Casualties")

plt.xlabel("Sex of Casualty")

plt.ylabel("Count")

plt.show()
```

Interpretation:

Accident casualty distribution varies across genders.

Casualty Type Distribution


```python
top_types = data['Casualty_Type'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_types.values,
    y=top_types.index,
    color='teal'
)

plt.title("Most Common Casualty Types")

plt.xlabel("Count")

plt.ylabel("Casualty Type")

plt.show()
```

Conclusion:
1. Traffic accident casualty patterns were analyzed successfully.
2. Casualty severity distribution was identified using visualization.
3. Young and middle-aged individuals were involved in more accidents.
4. Casualty distribution varied across gender and casualty type.
5. Data visualization helped identify important accident-related trends.
